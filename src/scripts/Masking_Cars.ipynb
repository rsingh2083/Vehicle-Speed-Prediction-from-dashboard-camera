```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
import csv
import json
```

```python
red_color = np.zeros([480,640,3],dtype=np.uint8)
red_color[:,:,0].fill(200)
red_color[:,:,1].fill(102)
red_color[:,:,2].fill(0)

green_color = np.zeros([480,640,3],dtype=np.uint8)
green_color[:,:,0].fill(0)
green_color[:,:,1].fill(255)
green_color[:,:,2].fill(163)

blue_color = np.zeros([480,640,3],dtype=np.uint8)
blue_color[:,:,0].fill(20)
blue_color[:,:,1].fill(0)
blue_color[:,:,2].fill(255)
    
# for i in range(1):
i=1
seg_path = "./Car_Segmented_Test/" + str(i)
org_path = "./testIMG/" + str(i)
# save_path = "./Car_Segmented_Test2_New/" + str(i)

orgimg=cv2.imread(org_path + ".jpg")

plt.imshow(orgimg)
plt.show()

img=cv2.imread(seg_path + ".png")

plt.imshow(img)
plt.show()

temp = np.bitwise_xor(img,red_color)
temp3 = np.nonzero(temp)
tempr = np.invert(temp)

tempr[temp3]=0

plt.imshow(np.invert(tempr))
plt.show()

temp = np.bitwise_xor(img,green_color)
temp3 = np.nonzero(temp)
tempg = np.invert(temp)

tempg[temp3]=0

temp = np.bitwise_xor(img,blue_color)
temp3 = np.nonzero(temp)
tempb = np.invert(temp)

tempb[temp3]=0

#     final = tempr+tempb+tempg

final = tempr

final = np.invert(final)

finalimg = np.bitwise_and(final,orgimg)

plt.imshow(finalimg)
plt.show()
# cv2.imwrite(save_path + ".jpg", finalimg)
```
